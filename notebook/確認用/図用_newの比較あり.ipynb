```python
TARGET = "BCR"
SAMPLE_SIZE = 100

if TARGET == "GMV":
    X_LIM_MAX = 30000
if TARGET == "BCR":
    X_LIM_MAX = 0.003

var_random_train = 0.002419416351
var_random_test = 0.002484038016

var_sep_dict_train = {
    "Kmeans": {
        "Proportional": 0.0022339732068361565,
        "Post": 0.002413541367850879,
        "Optimal": 0.0023185336277060722,
    },
    "Lasso-Kmeans": {
        "Proportional": 0.002147528290327496,
        "Post": 0.0024170723983618506,
        "Optimal": 0.002137733084439448,
    },
    "SFS-Kmeans-F": {
        "Proportional": 0.0023965269575522597,
        "Post": 0.0025037751739889853,
        "Optimal": 0.0024707103314737344,
    },
    "SFS-Kmeans-T": {
        "Proportional": 0.0021424617636450756,
        "Post": 0.0022165465181010254,
        "Optimal": 0.002404747233817939,
    },
}
var_sep_dict_test = {
    "Kmeans": {
        "Proportional": 0.002240754274896039,
        "Post": 0.002403027137480727,
        "Optimal": 0.0023376932508648073,
    },
    "Lasso-Kmeans": {
        "Proportional": 0.002134826132303113,
        "Post": 0.0024354894502956153,
        "Optimal": 0.0021744651387962605,
    },
    "SFS-Kmeans-F": {
        "Proportional": 0.002371173088107247,
        "Post": 0.00242462311156248,
        "Optimal": 0.0024268737321534763,
    },
    "SFS-Kmeans-T": {
        "Proportional": 0.002106356253391236,
        "Post": 0.0022216749212019223,
        "Optimal": 0.0022639513891532883,
    },
}

var_once_dict_train = {
    "Proportional": 0.002172079649210216,
    "Post": 0.0021984367223114986,
    "Optimal": 0.0022038247351564627,
}
var_once_dict_test = {
    "Proportional": 0.002208977577412837,
    "Post": 0.0022037350182825366,
    "Optimal": 0.002240192340746711,
}

var_once_new_dict_train = {
    "Proportional": 0.0021200060339798448,
    "Post": 0.0021902612706345735,
    "Optimal": 0.0022139077608629835,
}
var_once_new_dict_test = {
    "Proportional": 0.0021196869285215647,
    "Post": 0.002195918913363736,
    "Optimal": 0.0021079239108721002,
}


var_cuped_train = 0.0020816050222786675
var_cuped_test = 0.0021425977231300996

var_coss_train = 0.0022208390360000003
var_coss_test = 0.0022927017274999998
```

分散削減率の計算

```python
reduc_sep_dict_train = {}
reduc_sep_dict_test = {}
for method, d in var_sep_dict_train.items():
    reduc_sep_dict_train[method] = {}
    for allo, var in d.items():
        reduc_sep_dict_train[method][allo] = (1 - var / var_random_train) * 100
for method, d in var_sep_dict_test.items():
    reduc_sep_dict_test[method] = {}
    for allo, var in d.items():
        reduc_sep_dict_test[method][allo] = (1 - var / var_random_test) * 100

reduc_once_dict_train = {}
reduc_once_dict_test = {}
for allo, var in var_once_dict_train.items():
    reduc_once_dict_train[allo] = (1 - var / var_random_train) * 100
for allo, var in var_once_dict_test.items():
    reduc_once_dict_test[allo] = (1 - var / var_random_test) * 100

reduc_once_new_dict_train = {}
reduc_once_new_dict_test = {}
for allo, var in var_once_new_dict_train.items():
    reduc_once_new_dict_train[allo] = (1 - var / var_random_train) * 100
for allo, var in var_once_new_dict_test.items():
    reduc_once_new_dict_test[allo] = (1 - var / var_random_test) * 100

reduc_cuped_train = (1 - var_cuped_train / var_random_train) * 100
reduc_cuped_test = (1 - var_cuped_test / var_random_test) * 100

reduc_coss_train = (1 - var_coss_train / var_random_train) * 100
reduc_coss_test = (1 - var_coss_test / var_random_test) * 100
```

```python
def make_var_dict(
    var_random: float,
    var_cuped: float,
    var_coss: float,
    var_sep_dict: dict[str, dict[str, float]],
    var_once_dict: dict[str, float],
    var_once_new_dict: dict[str, float],
) -> dict:
    var_dict = {"Random": var_random, "CUPED": var_cuped, "COSS": var_coss}
    for method, _ in var_sep_dict.items():
        var_dict[method] = {
            "Optimal": var_sep_dict[method]["Optimal"],
            "Post": var_sep_dict[method]["Post"],
            "Proportional": var_sep_dict[method]["Proportional"],
        }
    var_dict["SFS-Kmeans-Var"] = {
        "Optimal": var_once_dict["Optimal"],
        "Post": var_once_dict["Post"],
        "Proportional": var_once_dict["Proportional"],
    }
    var_dict["SFS-Kmeans-Var_new"] = {
        "Optimal": var_once_new_dict["Optimal"],
        "Post": var_once_new_dict["Post"],
        "Proportional": var_once_new_dict["Proportional"],
    }

    return var_dict


def make_reduc_dict(
    reduc_cuped: float,
    reduc_coss: float,
    reduc_sep_dict: dict[str, dict[str, float]],
    reduc_once_dict: dict[str, float],
    reduc_once_new_dict: dict[str, float],
) -> dict:
    reduc_dict = {"CUPED": reduc_cuped, "COSS": reduc_coss}
    for method, _ in reduc_sep_dict.items():
        reduc_dict[method] = {
            "Optimal": reduc_sep_dict[method]["Optimal"],
            "Post": reduc_sep_dict[method]["Post"],
            "Proportional": reduc_sep_dict[method]["Proportional"],
        }
    reduc_dict["SFS-Kmeans-Var"] = {
        "Optimal": reduc_once_dict["Optimal"],
        "Post": reduc_once_dict["Post"],
        "Proportional": reduc_once_dict["Proportional"],
    }
    reduc_dict["SFS-Kmeans-Var_new"] = {
        "Optimal": reduc_once_new_dict["Optimal"],
        "Post": reduc_once_new_dict["Post"],
        "Proportional": reduc_once_new_dict["Proportional"],
    }

    return reduc_dict
```

```python
var_dict_train = make_var_dict(
    var_random=var_random_train,
    var_cuped=var_cuped_train,
    var_coss=var_coss_train,
    var_sep_dict=var_sep_dict_train,
    var_once_dict=var_once_dict_train,
    var_once_new_dict=var_once_new_dict_train,
)
var_dict_test = make_var_dict(
    var_random=var_random_test,
    var_cuped=var_cuped_test,
    var_coss=var_coss_train,
    var_sep_dict=var_sep_dict_test,
    var_once_dict=var_once_dict_test,
    var_once_new_dict=var_once_new_dict_test,
)

reduc_dict_train = make_reduc_dict(
    reduc_cuped=reduc_cuped_train,
    reduc_coss=reduc_coss_train,
    reduc_sep_dict=reduc_sep_dict_train,
    reduc_once_dict=reduc_once_dict_train,
    reduc_once_new_dict=reduc_once_new_dict_train,
)
reduc_dict_test = make_reduc_dict(
    reduc_cuped=reduc_cuped_test,
    reduc_coss=reduc_coss_test,
    reduc_sep_dict=reduc_sep_dict_test,
    reduc_once_dict=reduc_once_dict_test,
    reduc_once_new_dict=reduc_once_new_dict_test,
)
```
